```python
# ====================== 1. LOAD DATA ======================
df = pd.read_csv('/content/ecommerce_raw.csv')
print(f"Data awal: {df.shape[0]} baris")
print("Kolom:", df.columns.tolist())

text_column = 'full_text' # ← Ubah kalau kolom teksnya bukan 'text'

# ====================== 2. KAMUS ======================
kamus_alay = {
    # --- Singkatan umum ---
    "yg"          : "yang",
    "dgn"         : "dengan",
    "dg"          : "dengan",
    "dng"         : "dengan",
    "dlm"         : "dalam",
    "krn"         : "karena",
    "krna"        : "karena",
    "karna"       : "karena",
    "klo"         : "kalau",
    "kalo"        : "kalau",
    "tp"          : "tapi",
    "tpi"         : "tapi",
    "pdhl"        : "padahal",
    "utk"         : "untuk",
    "tdk"         : "tidak",
    "jgn"         : "jangan",
    "blm"         : "belum",
    "belom"       : "belum",
    "udah"        : "sudah",
    "udh"         : "sudah",
    "sdh"         : "sudah",
    "dah"         : "sudah",
    "uda"         : "sudah",
    "msh"         : "masih",
    "masi"        : "masih",
    "dpt"         : "dapat",
    "dapet"       : "dapat",
    "bs"          : "bisa",
    "bsa"         : "bisa",
    "jd"          : "jadi",
    "skrg"        : "sekarang",
    "skrng"       : "sekarang",
    "kmrn"        : "kemarin",
    "kemaren"     : "kemarin",
    "dr"          : "dari",
    "dri"         : "dari",
    "sm"          : "sama",
    "lg"          : "lagi",
    "lsg"         : "langsung",
    "lgsg"        : "langsung",
    "jg"          : "juga",
    "sy"          : "saya",
    "ak"          : "aku",
    "aq"          : "aku",
    "gue"         : "saya",
    "gua"         : "saya",
    "gw"          : "saya",
    "kk"          : "kakak",
    "kak"         : "kakak",
    "ka"          : "kakak",
    "km"          : "kamu",
    "kmu"         : "kamu",
    "lu"          : "kamu",
    "lo"          : "kamu",
    "loe"         : "kamu",
    "bgt"         : "banget",
    "bngt"        : "banget",
    "aja"         : "saja",
    "doang"       : "saja",
    "gt"          : "begitu",
    "gini"        : "begini",
    "gitu"        : "begitu",
    "gmn"         : "bagaimana",
    "gimana"      : "bagaimana",
    "knp"         : "kenapa",
    "ky"          : "kayak",
    "kaya"        : "seperti",
    "kayak"       : "seperti",
    "kayaknya"    : "sepertinya",
    "kyknya"      : "sepertinya",
    "emang"       : "memang",
    "emg"         : "memang",
    "mending"     : "lebih baik",
    "lbh"         : "lebih",
    "byk"         : "banyak",
    "org"         : "orang",
    "tgl"         : "tanggal",
    "hr"          : "hari",
    "bln"         : "bulan",
    "jm"          : "jam",
    "nomer"       : "nomor",
    "no"          : "nomor",
    "tlg"         : "tolong",
    "mhn"         : "mohon",
    "makasih"     : "terima kasih",
    "thx"         : "terima kasih",
    "tks"         : "terima kasih",
    "thanks"      : "terima kasih",
    "trims"       : "terima kasih",
    "terimakasih" : "terima kasih",
    "pake"        : "pakai",
    "dipake"      : "dipakai",
    "liat"        : "lihat",
    "bayangin"    : "bayangkan",
    "tau"         : "tahu",
    "tsb"         : "tersebut",
    "yaa"         : "ya",
    "nih"         : "ini",
    "tuh"         : "itu",
    "deh"         : "",
    "sih"         : "",
    "dong"        : "",
    "lho"         : "",

    # --- Negasi informal ---
    "gak"         : "tidak",
    "ngak"        : "tidak",
    "ga"          : "tidak",
    "gk"          : "tidak",
    "nggak"       : "tidak",
    "ngga"        : "tidak",
    "nggk"        : "tidak",
    "engga"       : "tidak",
    "enggak"      : "tidak",
    "kaga"        : "tidak",
    "kagak"       : "tidak",
    "gabisa"      : "tidak bisa",
    "gaada"       : "tidak ada",
    "gada"        : "tidak ada",
    "gapernah"    : "tidak pernah",
    "gausah"      : "tidak usah",
    "garespon"    : "tidak merespon",

    # --- Kata informal / gaul ---
    "bikin"       : "membuat",
    "nyoba"       : "mencoba",
    "ngirim"      : "mengirim",
    "nyampe"      : "sampai",
    "sampe"       : "sampai",
    "smpe"        : "sampai",
    "sampek"      : "sampai",
    "balik"       : "kembali",
    "balikin"     : "kembalikan",
    "bales"       : "balas",
    "dibales"     : "dibalas",
    "nanya"       : "bertanya",
    "ngadu"       : "mengadu",
    "ngalamin"    : "mengalami",
    "ngambil"     : "mengambil",
    "ngasih"      : "memberikan",
    "ngehubungin" : "menghubungi",
    "hubungin"    : "hubungi",
    "ngestuck"    : "tersangkut",
    "nyangkut"    : "tersangkut",
    "mesen"       : "memesan",
    "pesen"       : "memesan",
    "nyari"       : "mencari",
    "nemu"        : "menemukan",
    "nerima"      : "menerima",
    "masukin"     : "masukkan",
    "milih"       : "memilih",
    "kesel"       : "kesal",
    "nyusahin"    : "menyusahkan",
    "nangis"      : "menangis",
    "capek"       : "lelah",
    "lelet"       : "lambat",
    "lemot"       : "lambat",
    "ribet"       : "rumit",
    "becus"       : "mampu",
    "ancur"       : "hancur",
    "muter"       : "berputar",
    "mulu"        : "terus",
    "tbtb"        : "tiba-tiba",
    "gede"        : "besar",
    "benerin"     : "perbaiki",
    "dirubah"     : "diubah",
    "tetep"       : "tetap",
    "cepet"       : "cepat",
    "ketipu"      : "tertipu",
    "nipu"        : "menipu",
    "dibobol"     : "diretas",

    # --- Istilah e-commerce & medsos ---
    "tokped"      : "tokopedia",
    "toped"       : "tokopedia",
    "dmnya"       : "dm nya",
    "cdm"         : "chat dm",
    "infoin"      : "informasikan",
    "diinfoin"    : "diinformasikan",
    "lampirin"    : "lampirkan",
    "ajuin"       : "ajukan",
    "komplain"    : "keluhan",
    "complain"    : "keluhan",
    "komplen"     : "keluhan",
    "ilang"       : "hilang",
    "dateng"      : "datang",
    "dikasih"     : "diberikan",
    "otp"         : "kode otp",
    "ss"          : "screenshot",
    "sreenshot"   : "screenshot",
    "resi"        : "nomor resi",
    "resinya"     : "nomor resi",
    "ongkir"      : "ongkos kirim",
    "cod"         : "bayar di tempat",
    "sameday"     : "pengiriman hari yang sama",
    "scam"        : "penipuan",
    "modus"       : "modus penipuan",
    "ojol"        : "ojek online",
    "gopaylater"  : "gopay paylater",
    "spaylater"   : "shopee paylater",
    "tiktokshop"  : "tiktok shop",
    "sosmed"      : "media sosial",
    "reschedule"  : "jadwal ulang",
    "track"       : "lacak",
    "tracking"    : "pelacakan",
    "stuck"       : "tertahan",
    "follow up"   : "tindak lanjut",
    "update"      : "pembaruan",
    "refund"      : "pengembalian dana",
    "cancel"      : "batalkan",
    "dibatalin"   : "dibatalkan",
    "giveaway"    : "hadiah gratis",
    "voucher"     : "voucher diskon",
    "cashback"    : "pengembalian uang",
    "jnt"         : "j&t",
    "spx"         : "shopee express",
    "sicepat"     : "sicepat",
    "pickup"      : "penjemputan kurir",
    "dipickup"    : "dijemput kurir",
}
multiword_context = [
    # --- Keluhan pengiriman ---
    "tidak sampai",
    "belum sampai",
    "ga sampai",
    "tidak ada update",
    "ga ada update",
    "belum ada kabar",
    "tidak ada pergerakan",
    "ga ada pergerakan",
    "tidak bergerak",
    "ga bergerak",
    "stuck di",
    "tertahan di",
    "lama banget",
    "sangat lama",
    "terlalu lama",
    "nyasar ke",
    "nyangkut di",
    "paket hilang",
    "barang hilang",
    "kiriman hilang",
    "paket nyasar",
    "paket tidak bergerak",
    "nomor resi tidak valid",
    "resi tidak terlacak",
    "tidak bisa dilacak",
    "ga bisa dilacak",

    # --- Keluhan layanan ---
    "tidak direspon",
    "ga direspon",
    "belum dibalas",
    "ga dibalas",
    "tidak dibalas",
    "tidak bisa dihubungi",
    "susah dihubungi",
    "tidak ada solusi",
    "ga ada solusi",
    "tidak membantu",
    "ga membantu",
    "tidak berguna",
    "tidak jelas",
    "ga jelas",
    "tidak sesuai",
    "ga sesuai",
    "tidak aman",
    "ga aman",
    "ga nyaman",
    "tidak nyaman",
    "gak nyaman",
    "tidak profesional",
    "pelayanan buruk",
    "pelayanan jelek",
    "respon lambat",
    "respon lama",
    "tidak ada kepastian",
    "ga ada kejelasan",
    "tidak ada kejelasan",

    # --- Penipuan & keamanan ---
    "mengatasnamakan tokopedia",
    "mengatasnamakan shopee",
    "akun palsu",
    "akun resmi",
    "oknum tidak bertanggung jawab",
    "modus penipuan",
    "data bocor",
    "data dicuri",
    "akun dibobol",
    "transaksi tidak sah",
    "transaksi mencurigakan",
    "kode otp",
    "jangan berikan otp",
    "berhati hati penipuan",

    # --- Keluhan pembayaran & refund ---
    "uang belum kembali",
    "dana belum masuk",
    "refund belum cair",
    "saldo belum masuk",
    "pembayaran gagal",
    "tidak bisa bayar",
    "ga bisa bayar",
    "belum di refund",
    "uang hilang",
    "saldo berkurang",
    "dana berkurang",
    "terpotong otomatis",

    # --- Status pesanan ---
    "pesanan dibatalkan",
    "pesanan hilang",
    "barang tidak diterima",
    "barang belum diterima",
    "barang rusak",
    "barang salah",
    "barang tidak sesuai",
    "barang tidak sampai",
    "pesanan tidak sampai",
    "paket rusak",

    # --- Ekspresi kepuasan ---
    "terima kasih",
    "sangat membantu",
    "cepat sekali",
    "respon cepat",
    "pelayanan baik",
    "sangat puas",
    "tidak kecewa",
    "ga kecewa",
    "sudah beres",
    "sudah selesai",
    "sudah diterima",
    "sudah sampai",
    "terimakasih banyak",
    "makasih banyak",

    # --- Template balasan CS ---
    "mohon maaf",
    "mohon maaf atas",
    "atas ketidaknyamanannya",
    "ketidaknyamanan ini",
    "kami bantu cek",
    "kami segera cek",
    "boleh infoin",
    "silakan dm",
    "via dm",
    "melalui dm",
    "cek secara berkala",
    "pengecekan lebih lanjut",
    "lebih lanjut boleh",
    "live chat",
    "media sosial resmi",
    "follow up",
    "tindak lanjut",
    "stay safe",
    "sehat selalu",
    "jangan lupa mention",
    "infoin username dan nomor",
    "kirim nomor invoice",
    "di komen ini",

    # --- Konteks e-commerce ---
    "ongkos kirim",
    "gratis ongkir",
    "bebas ongkir",
    "bayar di tempat",
    "pengiriman hari yang sama",
    "pengembalian dana",
    "nomor resi",
    "nomor pesanan",
    "nomor invoice",
    "promo diskon",
    "voucher diskon",
    "cashback saldo",
    "shopee express",
    "amp express",
    "j&t express",
    "call center",
    "pusat bantuan",
    "lapor ke",
]

# ====================== 3. FUNGSI FULL PREPROCESSING ======================
def full_preprocessing(text):
  if not isinstance(text, str):
    return "", []

  text = text.lower().strip()
  text = re.sub(r'http\S+|www\S+|https\S+', '', text)
  text = re.sub(r'@\w+', '', text)
  text = re.sub(r'#\w+', '', text)
  text = re.sub(r'[^\w\s]', ' ', text)

  text = emoji.demojize(text, delimiters=("", ""))

  # Normalisasi slang
  words = text.split()
  words = [kamus_alay.get(word, word) for word in words]
  text = " ".join(words)

  # Penanganan negasi
  negasi = [
    # --- Formal ---
    "tidak",
    "bukan",
    "tak",
    "tanpa",
    "belum",
    "jangan",
    "tiada",
    "mustahil",
    "kurang",

    # --- Informal / gaul ---
    "ga",
    "gak",
    "ngga",
    "nggak",
    "ngak",
    "engga",
    "enggak",
    "gk",
    "kaga",
    "kagak",

    # --- Gabungan negasi (muncul di dataset) ---
    "gabisa",
    "gaada",
    "gada",
    "gapernah",
    "gausah",
    "garespon",

    # --- Singkatan ---
    "tdk",
    "blm",
    "jgn",
]
  tokens = text.split()
  for i in range(len(tokens)-1):
    if tokens[i] in negasi:
      tokens[i] = tokens[i] + "_" + tokens[i+1]
      tokens[i+1] = ""
  text = " ".join([t for t in tokens if t])

  # Repeated characters
  text = re.sub(r'(.)\1{2,}', r'\1\1', text)

  # Context-aware tokenization
  normalized_text = text
  for phrase in multiword_context:
    normalized_text = normalized_text.replace(phrase, phrase.replace("", "_"))

  tokens = word_tokenize(normalized_text)
  tokens = [t.replace("_", " ") for t in tokens if len(t) > 1]

  return normalized_text, tokens

# ====================== 4. TERAPKAN ======================
print("🔄️ Sedang memproses preprocessing...\n")

df[['clean_text', 'tokens_context']] = df[text_column].apply(lambda x: pd.Series(full_preprocessing(x)))

df.to_csv('ecommerce_preprocessed.csv', index=False)

# ====================== 5. HASIL ======================
print("PREPROCESSING SELESAI!\n")
print(f"Total data: {df.shape[0]} baris")
print("\nContoh hasil:")

for i in range(min(3, len(df))):
  print(f"\nOriginal : {df[text_column].iloc[i]}")
  print(f"Cleaned : {df['clean_text'].iloc[i]}")
  print(f"Tokens : {df['tokens_context'].iloc[i]}")

df['token_count'] = df['tokens_context'].apply(len)
print("\n Rata-rata jumlah token:", round(df['token_count'].mean(), 2))
print("File disimpan: **ecommerce_preprocessed.csv**")

import pandas as pd
import nltk
from nltk.tokenize import word_tokenize

# Download NLTK
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)

# ====================== LOAD DATA ======================
df = pd.read_csv('/content/ecommerce_preprocessed.csv')
print(f"Data loaded: {df.shape[0]} baris")

# ====================== LOAD & PERBAIKI KAMUS INSET ======================
pos_df = pd.read_csv('/content/positive.tsv', sep='\t', header=None,
names=['word', 'weight'])
neg_df = pd.read_csv('/content/negative.tsv', sep='\t', header=None,
names=['word', 'weight'])

# Penting: Ubah weight jadi tipe angka (float)
pos_df['weight'] = pd.to_numeric(pos_df['weight'], errors='coerce')
neg_df['weight'] = pd.to_numeric(neg_df['weight'], errors='coerce')

# Gabungkan jadi lexicon
lexicon = {}
for word, score in zip(pos_df['word'], pos_df['weight']):
  if pd.notna(score): # skip kalau ada NaN
    lexicon[word] = float(score)

for word, score in zip(neg_df['word'], neg_df['weight']):
  if pd.notna(score):
    lexicon[word] = float(score)

print(f" Kamus InSet berhasil dimuat: {len(lexicon)} kata")

# ====================== FUNGSI LABELING INSET ======================
def labeling_inset(text, lexicon):
  if not isinstance(text, str) or text.strip() == "":
    return 0.0, "Neutral", 0

  tokens = word_tokenize(text.lower())

  total_score = 0.0
  matched_words = 0

  for token in tokens:
    if token in lexicon:
      total_score += lexicon[token]
      matched_words += 1

  # Threshold InSet
  if total_score > 0:
    sentiment = "Positive"
  elif total_score < 0:
    sentiment = "Negative"
  else:
    sentiment = "Neutral"

  return total_score, sentiment, matched_words

# ====================== TERAPKAN LABELING ======================
print("🔄️ Sedang melakukan labeling InSet...\n")
results = df['clean_text'].apply(lambda x: labeling_inset(x,lexicon))
df['inset_score'] = results.apply(lambda x: x[0])
df['inset_sentiment'] = results.apply(lambda x: x[1])
df['matched_words'] = results.apply(lambda x: x[2])

# Simpan hasil
df.to_csv('ecommerce_inset_labeled_final.csv', index=False)

# ====================== HASIL ======================
print(" LABELING INSET SELESAI!\n")

print("Distribusi Sentimen:")
print(df['inset_sentiment'].value_counts())
print("\nPersentase:")
print(df['inset_sentiment'].value_counts(normalize=True).round(4)*100)

print("\nRata-rata InSet Score:", round(df['inset_score'].mean(), 3))
print("Rata-rata kata yang match dengan lexicon:",
round(df['matched_words'].mean(), 2))

print("\nContoh 5 hasil labeling:")
for i in range(min(5, len(df))):
  print(f"\nOriginal : {df['full_text'].iloc[i][:150]}...")
  print(f"Cleaned : {df['clean_text'].iloc[i]}")
  print(f"Score :{df['inset_score'].iloc[i]:.2f} → {df['inset_sentiment'].iloc[i]}")

print("\n💾 File disimpan: **ecommerce_inset_labeled_final.csv**")
```

**Menampilkan Mention dan Hastag**

```python
import pandas as pd
import re

# Baca data
df = pd.read_csv("ecommerce_raw.csv")

# Pastikan kolom full_text tidak kosong
df = df.dropna(subset=["full_text"]).copy()

def format_tweet(row):
    text = str(row["full_text"]).lower()

    # Ambil username dari kolom username
    username = str(row["username"]).strip()

    # Jika username kosong / NaN
    if username == "nan" or username == "":
        username = "@user"
    elif not username.startswith("@"):
        username = "@" + username

    # Ambil semua hashtag dari tweet
    hashtags = re.findall(r"#\w+", text)

    # Hilangkan hashtag lama dari isi tweet
    text_clean = re.sub(r"#\w+", "", text)

    # Hilangkan mention lama dari isi tweet
    text_clean = re.sub(r"@\w+", "", text_clean)

    # Hilangkan URL
    text_clean = re.sub(r"http\S+|www\.\S+", "", text_clean)

    # Rapikan spasi
    text_clean = re.sub(r"\s+", " ", text_clean).strip()

    # Hilangkan hashtag duplikat
    hashtags = list(dict.fromkeys(hashtags))

    # Gabungkan format:
    # @username### #hashtag1 #hashtag2 isi tweet
    hasil = f"{username}###"

    if hashtags:
        hasil += " " + " ".join(hashtags)

    if text_clean:
        hasil += " " + text_clean

    return hasil

# Buat kolom baru
df["tweet_lda"] = df.apply(format_tweet, axis=1)

# Lihat hasil
print(df[["username", "full_text", "tweet_lda"]].head(10).to_string())

# Simpan hasil
df.to_csv(
    "ecommerce_format_mention_hashtag.csv",
    index=False,
    encoding="utf-8-sig"
)

print("\nFile berhasil dibuat:")
print("ecommerce_format_mention_hashtag.csv")
```

**Dictionary Corpus**

```python
from gensim import corpora
import ast
import pandas as pd # Import pandas to reload the dataframe

# Load the preprocessed dataframe which contains 'tokens_context'
df = pd.read_csv('ecommerce_preprocessed.csv')

# Convert the 'tokens_context' column from string representation of list to actual list
df['tokens_context'] = df['tokens_context'].apply(ast.literal_eval)

dictionary = corpora.Dictionary(df["tokens_context"])

# Filter kata yang terlalu jarang atau terlalu sering muncul
dictionary.filter_extremes(
    no_below=2,
    no_above=0.80,
    keep_n=5000
)

corpus = [dictionary.doc2bow(tokens) for tokens in df["tokens_context"]]

print("\nJumlah kata unik setelah filter:", len(dictionary))
print("Jumlah dokumen untuk LDA:", len(corpus))
```

**Vektorisasi : Bag Of Words**

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

print("\n" + "="*55)
print("BAGIAN 3: VEKTORISASI BAG OF WORDS")
print("="*55)

# PENTING: LDA menggunakan CountVectorizer (frekuensi kata mentah)
# BUKAN TF-IDF! TF-IDF akan mengacaukan asumsi probabilistik LDA
cv = CountVectorizer(
    max_features = 300, # ambil 300 kata paling sering
    min_df = 2, # kata harus muncul di min. 2 dokumen
    max_df = 0.90, # abaikan kata yang ada di > 90% dokumen
    ngram_range = (1, 1), # unigram saja untuk LDA
)

# Fill any NaN values in 'clean_text' with empty strings
df['clean_text'] = df['clean_text'].fillna('')
X = cv.fit_transform(df['clean_text'])
terms = cv.get_feature_names_out()

print(f"\n Vektorisasi selesai!")
print(f" Dimensi matriks BoW : {X.shape} (dokumen × kata)")
print(f" Jumlah fitur kata : {len(terms)}")
print(f" Top 30 kata paling sering:")

# Hitung frekuensi total setiap kata
word_freq = np.array(X.sum(axis=0)).flatten()
top30_idx = word_freq.argsort()[::-1][:30]
print(f" {', '.join([terms[i] for i in top30_idx])}")
```

**Mencari K Optimal - Elbow Perplexity**

```python
from sklearn.decomposition import LatentDirichletAllocation
import numpy as np
import matplotlib.pyplot as plt

print("\n" + "="*55)
print("BAGIAN 4: MENCARI K OPTIMAL — ELBOW PERPLEXITY")
print("="*55)
K_range = range(2, 11)
perplexities = []
log_likelihoods = []
print("\n K | Perplexity | Log-Likelihood")
print(" " + "-"*38)
for k in K_range:
    lda_tmp = LatentDirichletAllocation(
        n_components = k,
        max_iter = 15,
        learning_method = 'online',
        random_state = 42
    )
    lda_tmp.fit(X)
    perp = lda_tmp.perplexity(X)
    logl = lda_tmp.score(X)
    perplexities.append(perp)
    log_likelihoods.append(logl)
    print(f" K={k:2d} | {perp:10.2f} | {logl:12.2f}")

best_k = list(K_range)[np.argmin(perplexities)]
print(f"\n K terbaik (perplexity terendah) = {best_k}")
# Plot Elbow
fig, axes = plt.subplots(1, 2, figsize=(13, 4))
axes[0].plot(list(K_range), perplexities, 'o-', color='#7C3AED', lw=2.5, ms=9)
axes[0].axvline(x=best_k, color='red', linestyle='--', alpha=0.7, label=f'K\noptimal={best_k}')
axes[0].set_title('Perplexity vs K (↓ semakin baik)', fontsize=13,
fontweight='bold')
axes[0].set_xlabel('Jumlah Topik (K)')
axes[0].set_ylabel('Perplexity')
axes[0].legend(); axes[0].grid(alpha=0.3)
axes[1].plot(list(K_range), log_likelihoods, 's-', color='#0891B2', lw=2.5,
ms=9)
axes[1].axvline(x=best_k, color='red', linestyle='--', alpha=0.7, label=f'K\noptimal={best_k}')
axes[1].set_title('Log-Likelihood vs K (↑ semakin baik)', fontsize=13,
fontweight='bold')
axes[1].set_xlabel('Jumlah Topik (K)')
axes[1].set_ylabel('Log-Likelihood')
axes[1].legend(); axes[1].grid(alpha=0.3)

plt.suptitle('Evaluasi LDA — Menentukan K Optimal | ecommerce_raw.csv',
fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('01_lda_elbow.png', dpi=150, bbox_inches='tight')
plt.show()
print(" Grafik disimpan: 01_lda_elbow.png")
```

**Fit Model LDA Final**

```python
print("\n" + "="*55)
print("BAGIAN 5: FIT MODEL LDA FINAL")
print("="*55)
# bisa diganti n_components sesuai hasil evaluasi di atas
N_TOPICS = best_k # atau set manual, misal: N_TOPICS = 5

lda = LatentDirichletAllocation(
n_components = N_TOPICS,
max_iter = 30, # iterasi lebih banyak untuk konvergen
learning_method = 'online', # online variational Bayes (lebih cepat)
doc_topic_prior = None, # alpha = 1/K (default)
topic_word_prior= None, # beta = 1/K (default)
random_state = 42,
)
lda.fit(X)
print(f"\n LDA selesai dilatih!")
print(f" Jumlah topik : {lda.n_components}")
print(f" Log-likelihood : {lda.score(X):.2f}")
print(f" Perplexity : {lda.perplexity(X):.2f}")
```

**Hasil Topik LDA**

```python
print("\n" + "="*55)
print("BAGIAN 6: HASIL TOPIK LDA")
print("="*55)
N_TOP_WORDS = 15
print(f"\n{'='*60}")
print(f" TOPIK-TOPIK yang ditemukan LDA dari ecommerce_raw.csv")
print(f"{'='*60}")
for topic_idx, topic_comp in enumerate(lda.components_):
    top_indices = topic_comp.argsort()[::-1][:N_TOP_WORDS]
    top_words = [terms[i] for i in top_indices]
    top_weights = [topic_comp[i] for i in top_indices]
    print(f"\n TOPIK {topic_idx}:")
    print(f" Top kata : {', '.join(top_words[:10])}")
    print(f" Bobot : {', '.join([f'{w:.1f}' for w in top_weights[:10]])}")
```

**Assign Topik Setiap Tweet**

```python
print("\n" + "="*55)
print("BAGIAN 7: ASSIGN TOPIK KE SETIAP TWEET")
print("="*55)
doc_topic_matrix = lda.transform(X) # shape: (n_docs, n_topics)
df['dominant_topic'] = doc_topic_matrix.argmax(axis=1)
df['topic_prob'] = doc_topic_matrix.max(axis=1)
print(f"\n Distribusi topik dominan:")
print(df['dominant_topic'].value_counts().sort_index().to_string())
print(f"\n Contoh tweet per topik:")
for k in range(N_TOPICS):
    subset = df[df['dominant_topic'] == k].sort_values('topic_prob',
        ascending=False)
    print(f"\n ── Topik {k} (n={len(subset)}, avg_prob={subset['topic_prob'].mean():.3f}) ──")
    for _, row in subset.head(2).iterrows():
        print(f" → [{row['topic_prob']:.2f}] {str(row['full_text'])[:85]}...")
```

# **5. NER Rule-Based**

```python
import pandas as pd
import re
import string

# 1. Load Data
df = pd.read_csv("ecommerce_inset_labeled_final.csv")

# 2. Kamus untuk Feature Assignment
OPOS_DICTIONARY = ["ketua", "direktur", "gubernur", "presiden", "menteri", "admin", "cs", "min"]
POS_DICT = {
    "paketnya": "NOUN", "udah": "ADV", "ada": "VERB", "pergerakan": "NOUN",
    "belum": "ADV", "punyaku": "NOUN", "juga": "ADV", "dari": "PREP",
    "tgl": "NOUN", "ga": "ADV", "update": "VERB", "sampai": "PREP", "skrg": "ADV",
    "kak": "NOUN", "shopee": "NOUN", "dm": "NOUN", "ku": "PRON", "kemarin": "ADV",
    "iyaa": "ADV", "udahh": "ADV", "samaa": "ADV", "bgtt": "ADV", "bayangin": "VERB",
    "dr": "PREP", "hari": "NOUN", "posisi": "NOUN", "nya": "PRON", "dimana": "PREP",
    "mana": "PREP", "brg": "NOUN", "ny": "PRON", "mau": "ADV", "di": "PREP",
    "pake": "VERB", "buat": "PREP", "konseran": "NOUN", "chat": "VERB", "sekarang": "ADV"
}

all_results = []

# 3. Proses Tokenisasi dan Feature Assignment untuk semua baris
for idx, row in df.iterrows():
    text = str(row['full_text'])
    if pd.isna(text) or text == 'nan': continue

    # Pembersihan teks dari mention dan url
    text_clean = re.sub(r'@\w+', '', text)
    text_clean = re.sub(r'http\S+', '', text_clean).strip()

    # Memisahkan kata dan tanda baca menggunakan regex
    raw_tokens = re.findall(r'\w+|[^\w\s]', text_clean)

    sentence_tokens = []

    for token in raw_tokens:
        if token.strip() == '' or len(token.encode('utf-8')) > 3: # Skip empty or simple emojis
            # further filter for emojis by checking character category if needed,
            # but simple length check on bytes avoids most complex emojis here.
            pass

        kind = ""
        contextual = ""
        morph = []
        pos = ""

        # Penentuan Token Kind
        if token.isdigit():
            kind = "NUM"
            morph.append("Digit")
        elif any(p in token for p in string.punctuation):
            if token in ['?', '.', '!']: kind = "EPUNC"
            elif token in [',']: kind = "OPUNC"
            elif token in ['(', ')', '-', '/']: kind = "SPUNC"
            else: kind = "PUNC"
        else:
            # Jika hanya huruf atau alfanumerik
            kind = "WORD" if token.isalnum() else "PUNC"

            # Morphological Features
            if token.istitle():
                morph.append("TitleCase")
                morph.append("CapStart")
            elif token.isupper() and len(token) > 1:
                morph.append("UpperCase")
                morph.append("CapStart")
            elif token.islower():
                morph.append("LowerCase")

            if any(c.isdigit() for c in token) and any(c.isalpha() for c in token):
                morph.append("CharDigit")

            # Contextual Features
            if token.lower() in OPOS_DICTIONARY:
                contextual = "OPOS"

            # POS Features
            if token.lower() in POS_DICT:
                pos = POS_DICT[token.lower()]
            else:
                pos = "OOV"

        # Hindari memasukkan emoji murni ke tabel (opsional pembersihan)
        if not re.match(r'^[\w\W]*$', token):
            continue

        token_dict = {
            "Sentence_ID": idx + 1,
            "Token string": token,
            "Token kind": kind,
            "Contextual features": contextual,
            "Morphological features": ", ".join(morph) if morph else "",
            "Part-of-speech features": pos,
            "Type of Named Entity": '""' # Default value sebelum rules
        }
        sentence_tokens.append(token_dict)

    # 4. Implementasi Logika IF THEN (Rule Assignment)
    for i in range(len(sentence_tokens) - 1):
        # IF Token[i].Kind="WORD" and Token[i].OPOS
        cond_1 = (sentence_tokens[i]["Token kind"] == "WORD" and
                  sentence_tokens[i]["Contextual features"] == "OPOS")

        # and Token[i+1].Kind=”WORD” and Token[i+1].UpperCase and Token[i+1].OOV
        cond_2 = (sentence_tokens[i+1]["Token kind"] == "WORD" and
                  "UpperCase" in sentence_tokens[i+1]["Morphological features"] and
                  sentence_tokens[i+1]["Part-of-speech features"] == "OOV")

        # THEN Token[i+1].NE = "ORGANIZATION"
        if cond_1 and cond_2:
            sentence_tokens[i+1]["Type of Named Entity"] = "ORGANIZATION"

    all_results.extend(sentence_tokens)

# 5. Buat DataFrame dan simpan ke CSV
final_df = pd.DataFrame(all_results)
output_path = "ecommerce_feature_assignment.csv"
final_df.to_csv(output_path, index=False)

# 6. Tampilkan 5 kalimat (Sentence_ID 1 s/d 5)
display_df = final_df[final_df['Sentence_ID'] <= 5].drop(columns=['Sentence_ID'])
print(display_df.head(25))
```

# **6. Menghitung FTC**

```python
from collections import defaultdict
import itertools
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab', quiet=True)

def FTC(D, minsup):
    SelectedTermSets = set()
    n = len(D)
    min_count = int(minsup * n)

    def DetermineFrequentTermsets(database_tokens, min_support_count):
        term_counts = defaultdict(int)
        frequent_term_sets = set()
        for doc_tokens in database_tokens:
            unique_doc_tokens = list(set(doc_tokens))
            for term in unique_doc_tokens:
                term_counts[frozenset({term})] += 1
            for pair in itertools.combinations(unique_doc_tokens, 2):
                term_counts[frozenset(pair)] += 1
        for term_set, count in term_counts.items():
            if count >= min_support_count:
                frequent_term_sets.add(term_set)
        return frequent_term_sets

    def get_documents_covered_by_term_set(term_set_to_check, D_original_tokens):
        covered_docs = set()
        for doc_id, doc_tokens in enumerate(D_original_tokens):
            if term_set_to_check.issubset(set(doc_tokens)):
                covered_docs.add(doc_id)
        return covered_docs

    RemainingTermSets = DetermineFrequentTermsets(D, min_count)
    uncovered_documents = set(range(n))

    while len(uncovered_documents) != 0:
        if not RemainingTermSets: break
        best_candidate = None
        max_new_coverage = -1
        for candidate_set in RemainingTermSets:
            candidate_coverage = get_documents_covered_by_term_set(candidate_set, D)
            new_coverage = len(candidate_coverage.intersection(uncovered_documents))
            if new_coverage > max_new_coverage:
                max_new_coverage = new_coverage
                best_candidate = candidate_set
        if best_candidate is None or max_new_coverage == 0: break
        SelectedTermSets.add(best_candidate)
        RemainingTermSets.remove(best_candidate)
        uncovered_documents.difference_update(get_documents_covered_by_term_set(best_candidate, D))

    total_covered = set()
    for ts in SelectedTermSets:
        total_covered.update(get_documents_covered_by_term_set(ts, D))
    return SelectedTermSets, total_covered

try:
    df_ftc = pd.read_csv('ecommerce_preprocessed.csv')
    # Fix: Drop rows with NaN in critical columns to prevent float errors
    df_ftc = df_ftc.dropna(subset=['clean_text'])
    df_ftc['tokens'] = df_ftc['clean_text'].astype(str).apply(word_tokenize)

    actual_D = df_ftc['tokens'].tolist()
    actual_minsup = 0.05

    print("--- Menjalankan FTC ---")
    selected_sets, covered_docs = FTC(actual_D, actual_minsup)

    print("\nSelected Term Sets:")
    for ts in selected_sets:
        print(f" - {set(ts)}")
    print(f"\nTotal Documents Covered: {len(covered_docs)} dari {len(actual_D)}")
except Exception as e:
    print(f"Error: {e}")
```

```python
from collections import Counter
import ast # Untuk mengonversi string list menjadi list

all_terms = []

# Periksa tipe data di df['tokens']
# Jika berupa string, ubah menjadi list menggunakan ast.literal_eval
# Jika sudah list, langsung gunakan
if isinstance(df['tokens'].iloc[0], str):
    # Mengonversi string representasi list menjadi list sebenarnya
    df['tokens'] = df['tokens'].apply(ast.literal_eval)

for tokens in df["tokens"]:
    all_terms.extend(tokens)

term_frequency = Counter(all_terms)

df_frequency = pd.DataFrame(
    term_frequency.items(),
    columns=["term", "frequency"]
)

df_frequency = df_frequency.sort_values(
    by="frequency",
    ascending=False
).reset_index(drop=True)

print("\nTop 20 term paling sering muncul:")
print(df_frequency.head(20))
```

**Menghitung Entropy**

```python
import math # Import the math module
total_documents = len(df)

term_entropy = {}

for term in term_frequency.keys():

    # Tweet yang memuat term tersebut
    doc_indices = []

    for index, tokens in enumerate(df["tokens"]):
        if term in tokens:
            doc_indices.append(index)

    jumlah_muncul = len(doc_indices)

    # Karena setiap tweet dianggap satu kemunculan term,
    # probabilitasnya seragam pada tweet yang memuat term.
    if jumlah_muncul > 0:
        probability = 1 / jumlah_muncul
        entropy = -jumlah_muncul * probability * math.log2(probability)
    else:
        entropy = 0

    term_entropy[term] = entropy

df_entropy = pd.DataFrame(
    term_entropy.items(),
    columns=["term", "entropy"]
)

# Gabungkan frekuensi + entropy
df_term = pd.merge(
    df_frequency,
    df_entropy,
    on="term",
    how="left"
)

df_term = df_term.sort_values(
    by=["frequency", "entropy"],
    ascending=[False, True]
).reset_index(drop=True)

print("\nTop term dengan frequency dan entropy:")
print(df_term.head(20))
```

```python
minimum_frequency = 3

frequent_terms = df_term[
    df_term["frequency"] >= minimum_frequency
].copy()

print("\nJumlah frequent terms:", len(frequent_terms))
```

```python
from itertools import combinations # Import combinations for pairwise term processing

term_documents = {}

for term in frequent_terms["term"]:
    term_documents[term] = set(
        df.index[
            df["tokens"].apply(lambda tokens: term in tokens)
        ]
    )

overlap_results = []

frequent_term_list = frequent_terms["term"].tolist()

for term_a, term_b in combinations(frequent_term_list, 2):

    docs_a = term_documents[term_a]
    docs_b = term_documents[term_b]
    intersection = len(docs_a.intersection(docs_b)) # Corrected indentation
    min_docs = min(len(docs_a), len(docs_b))

    if min_docs > 0:
        overlap_score = intersection / min_docs
    else:
        overlap_score = 0

    overlap_results.append({
        "term_1": term_a,
        "term_2": term_b,
        "frequency_term_1": len(docs_a),
        "frequency_term_2": len(docs_b),
        "co_occurrence": intersection,
        "overlap": round(overlap_score, 4)
    })

df_overlap = pd.DataFrame(overlap_results)

df_overlap = df_overlap.sort_values(
    by="overlap",
    ascending=False
).reset_index(drop=True)

print("\nTop 20 pasangan term dengan overlap tertinggi:")
print(df_overlap.head(20))
```

```python
from collections import defaultdict

# 1. Bangun Adjacency List untuk pencarian cepat
print("Membangun peta hubungan kata (Adjacency List)...")
adj_list = defaultdict(list)

# Filter overlap yang memenuhi threshold terlebih dahulu
filtered_overlap = df_overlap[df_overlap['overlap'] >= 0.5]

for _, row in filtered_overlap.iterrows():
    t1, t2 = row['term_1'], row['term_2']
    adj_list[t1].append(t2)
    adj_list[t2].append(t1)

# 2. Jalankan algoritma pengelompokan (BFS/DFS)
clusters_opt = []
visited_terms = set()

print(f"Memulai clustering untuk {len(frequent_term_list)} term...")

for term in frequent_term_list:
    if term in visited_terms:
        continue

    # Gunakan penelusuran graf sederhana untuk menemukan satu kelompok
    current_cluster = {term}
    queue = [term]
    visited_terms.add(term)

    while queue:
        curr = queue.pop(0)
        for neighbor in adj_list.get(curr, []):
            if neighbor not in visited_terms:
                visited_terms.add(neighbor)
                current_cluster.add(neighbor)
                queue.append(neighbor)

    clusters_opt.append(sorted(list(current_cluster)))

print(f"Selesai! Berhasil membuat {len(clusters_opt)} cluster.")
# Tampilkan contoh 5 cluster pertama
for i, c in enumerate(clusters_opt[:5]):
    print(f"Cluster {i+1}: {c}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

if not df_overlap.empty:
    # 1. Pilih Top N terms (misal 50) untuk performa dan keterbacaan
    top_n = 50
    top_terms = frequent_term_list[:top_n]

    # 2. Filter data overlap hanya untuk top terms tersebut
    df_overlap_top = df_overlap[
        df_overlap['term_1'].isin(top_terms) &
        df_overlap['term_2'].isin(top_terms)
    ]

    # 3. Bangun matriks secara efisien menggunakan pivot
    # Ini jauh lebih cepat daripada menggunakan loop manual
    matrix = df_overlap_top.pivot(index='term_1', columns='term_2', values='overlap').fillna(0)

    # 4. Pastikan matriks memiliki semua top_terms dan simetris
    matrix = matrix.reindex(index=top_terms, columns=top_terms, fill_value=0)

    # Gunakan operasi NumPy untuk membuat matriks simetris (A = A + A.T)
    full_matrix = np.maximum(matrix.values, matrix.values.T)

    # Isi diagonal dengan 1.0 (overlap term dengan dirinya sendiri)
    np.fill_diagonal(full_matrix, 1.0)

    # 5. Visualisasi
    plt.figure(figsize=(16, 12))
    sns.heatmap(
        full_matrix,
        xticklabels=top_terms,
        yticklabels=top_terms,
        cmap='viridis',
        annot=False,
        linewidths=.1,
        linecolor='lightgray'
    )
    plt.title(f'Optimized Heatmap of Term Overlap (Top {top_n} Terms)', fontsize=16)
    plt.xlabel('Term', fontsize=12)
    plt.ylabel('Term', fontsize=12)
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
else:
    print("df_overlap kosong, tidak dapat membuat heatmap.")
```
